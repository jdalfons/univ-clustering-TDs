# K-Means avec "scikit-learn"

```python
#outil standardisation des variables
from sklearn.preprocessing import StandardScaler
std = StandardScaler()

#centrage-réduction
ZVA = std.fit_transform(XVA)

#ZDA est une matrice numpy
print(type(ZVA))
```

```python
#à transformer en data frame pandas
ZVA = pandas.DataFrame(ZVA,index=XVA.index,columns=XVA.columns)
ZVA.describe()
```

```python
#si on utilise numpy
import numpy
numpy.std(ZVA,axis=0,ddof=0)
```

```python
#k-means avec 3 groupes dans un premier temps
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3,n_init=10,random_state=0)
kmeans.fit(ZVA)
```

```python
#inertie résiduelle WSS
WSS = kmeans.inertia_
WSS
```

```python
#classes d'appartenance des individus
kmeans.labels_
```

```python
#groupe d'appartenance associé aux noms des modèles
pandas.DataFrame(kmeans.labels_,index=XVA.index)
```

```python
#effectifs par classe
numpy.unique(kmeans.labels_,return_counts=True)
```

```python
#récupération des effectifs dans un vecteur
effectifs = numpy.unique(kmeans.labels_,return_counts=True)[1]
print(effectifs)
```

```python
#centres de classes
kmeans.cluster_centers_
```

```python
#affichage plus sympathique
dfTemp = pandas.DataFrame(kmeans.cluster_centers_,index=['0','1','2'],columns=XVA.columns)
dfTemp
```

```python
#calcul du BSS
BSSk = effectifs * numpy.sum(kmeans.cluster_centers_**2,axis=1)
BSSk
```

```python
#et donc BSS
BSS = numpy.sum(BSSk)
BSS
```

```python
#addition BSS + WSS devrait être égal à TSS
BSS + WSS
```

```python
#calcul direct du TSS -- OK
#nb. d'obs * nb. de variables
TSS = ZVA.shape[0]*ZVA.shape[1]
TSS
```

```python
#inertie expliquée par la partition en 3 groupes (en pourcentage)
R2 = BSS/TSS
R2*100
```

# Détermination du nombre de classes

```python
#rendre silencieux les warnings à partir d'ici
import warnings
warnings.filterwarnings(action='ignore')

#courbe pour identifier le nombre de groupes - de 1 à 10
kWSS = numpy.zeros(11)
#nombre de clusters = 1 => WSS = TSS
kWSS[1] = TSS
#boucler
for k in range(2,11):
    km = KMeans(n_clusters=k,n_init=10)
    km.fit(ZVA)
    kWSS[k] = km.inertia_

#affichage de contrôle
kWSS
```

```python
#courbe
import matplotlib.pyplot as plt
plt.title("WSS")
plt.xlabel("# of clusters")
plt.plot(numpy.arange(1,11),kWSS[1:],marker='X',linestyle='solid')
plt.show()
```

```python
#et si on utilise yellowbrick maintenant ?
from yellowbrick.cluster import KElbowVisualizer
visualizer = KElbowVisualizer(KMeans(n_init=10),k=(1,11),metric='distortion')
visualizer.fit(ZVA)
visualizer.show()
```

```python
#et avec de critère de Calinski-Harabasz
visualizer = KElbowVisualizer(KMeans(n_init=10),k=(2,11),metric='silhouette')
visualizer.fit(ZVA)
visualizer.show()
```

# Interprétation avec les variables supplémentaires

```python
#data frame avec les variables supplémentaires et l'appartenance aux groupes
XSupp = D[['prix','origine','carburant','type4X4']]
XSupp['clusters'] = kmeans.labels_
XSupp.info()
```

```python
#calculer le prix moyen des véhicules de chaque groupe
pandas.pivot_table(XSupp,index='clusters',values='prix',aggfunc='mean')
```

```python
#pour le calcul du KHI-2 du tableau de contingence
from scipy.stats import chi2_contingency

#calculer les V de Cramer de chaque variable supplémentaire
#inspiré de "https://www.kaggle.com/code/chrisbss1/cramer-s-v-correlation-matrix/notebook"
for var in ['origine','carburant','type4X4']:
    crosstab = numpy.array(pandas.crosstab(XSupp.clusters,XSupp[var],rownames=None,colnames=None))
    stat = chi2_contingency(crosstab)[0]
    obs = numpy.sum(crosstab)
    mini = min(crosstab.shape)-1
    v = numpy.sqrt(stat/(obs*mini))
    print("v de Cramer de {} = {}".format(var,v))
```

```python
#ou plus simplement en scrutant la doc en ligne de scipy
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.contingency.association.html
from scipy.stats.contingency import association
for var in ['origine','carburant','type4X4']:
    crosstab = numpy.array(pandas.crosstab(XSupp.clusters,XSupp[var],rownames=None,colnames=None))
    v = association(crosstab,method='cramer')
    print("v de Cramer de {} = {}".format(var,v))
```

```python
#proportion de type4X4 = oui dans chaque cluster
pandas.crosstab(XSupp.clusters,XSupp.type4X4,normalize='index')
```

```python
#répartition de type4X4 = oui selon les clusters
pandas.crosstab(XSupp.clusters,XSupp.type4X4,normalize='columns')
```
