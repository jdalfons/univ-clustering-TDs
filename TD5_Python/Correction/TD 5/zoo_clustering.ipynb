# Importation et préparation des données

```python
#changement de dossier
import os
os.chdir("C:/Users/ricco/Desktop/demo")

#chargement des données
import pandas
zoo = pandas.read_excel("zoo_dataset.xlsx")
zoo.head()
```

```python
#information
zoo.info()
```

```python
#isoler les variables actives
X = zoo[zoo.columns[:-1]]
X.info()
```

```python
#codage O/1
M = pandas.get_dummies(X,drop_first=False)
print(type(M))
M.info()
```

```python
#effectifs par modalité
M.describe()
```

```python
#matrice M en numpy
MN = M.values
print(MN)
```

```python
#numpy
import numpy
#fonction indice de Dice au carré
def funSqDice(col1,col2):
    return(0.5*numpy.sum((col1-col2)**2))
```

```python
#calcul de la matrice de distance -- Indice de Dice
Dice = numpy.zeros(shape=(M.shape[1],M.shape[1]))

#remplissage
for k1 in range(MN.shape[1]):
    for k2 in range(MN.shape[1]):
        Dice[k1,k2] = funSqDice(MN[:,k1],MN[:,k2])

#passer en racine carrée
Dice = numpy.sqrt(Dice)
print(Dice[:5,:5]) #affichage d'une partie seulement
```

```python
#vectoriser pour la CAH
from scipy.spatial.distance import squareform
vDice = squareform(Dice)
print(vDice)
```

```python
#CAH, ward
from scipy.cluster.hierarchy import ward
cah = ward(vDice)
print(cah)
```

```python
#hauteur d'agrégation
hauteur = numpy.flip(numpy.sort(cah[:,2]))
print(hauteur)
```

```python
#courbe
import matplotlib.pyplot as plt
plt.title("Paliers d'agregation")
plt.plot(numpy.arange(1,hauteur.shape[0]+1,1),hauteur)
plt.show()
```

```python
#dendrogramme
from scipy.cluster.hierarchy import dendrogram
plt.title("CAH")
dendrogram(cah,labels=M.columns,orientation='left',color_threshold=0)
plt.show()
```

```python
#matérialisation de 6 classes
plt.title("CAH - 6 classes")
dendrogram(cah,labels=M.columns,orientation='left',color_threshold=5.0)
plt.show()
```

```python
#découpage en 6 classes
from scipy.cluster.hierarchy import fcluster
groupes = fcluster(cah,t=5.0,criterion='distance')
print(groupes)
```

```python
#effectifs par groupe
print(numpy.unique(groupes,return_counts=True))
```

```python
#dummies de la variable "genre"
genres = pandas.get_dummies(zoo.genre,drop_first=False)
genres.info()
```

```python
#calcul de la moyenne de l'indice de Dice
#de mammal avec les groupes
for g in range(1,7):
    #isoler les variables corresp.
    tmpM = M[M.columns[groupes==g]]
    print(tmpM.columns)
    #valeur indice de Dice
    dices = tmpM.apply(axis=0,func=lambda x:funSqDice(x,genres.mammal))
    print(numpy.mean(dices))
```

```python
#lancer l'ACM (analyse des correspondances multiples)
from fanalysis.mca import MCA
acm = MCA(row_labels=X.index,var_labels=X.columns)
acm.fit(X.values)
```

```python
#valeurs propres
print(acm.eig_)
```

```python
#représentation des valeurs propres
acm.plot_eigenvalues("percentage")
```

```python
#coordonnées des variables
acm.col_topandas()
```

```python
#coordonnées des variables dans le plan
coordVar = acm.col_topandas().iloc[:,:2]
coordVar.columns = ["F1","F2"]
print(coordVar)
```

```python
#ajouter le groupe d'appartenance
coordVar['grp'] = groupes
print(coordVar)
```

```python
#graphique
import seaborn as sns
sns.scatterplot(data=coordVar,x="F1",y="F2",hue="grp",palette=sns.color_palette()[:6])
for i,v in enumerate(M.columns):
    plt.annotate(v,(coordVar.iloc[i,0],coordVar.iloc[i,1]),color=sns.color_palette()[coordVar.grp[i]-1])
plt.show()
```
