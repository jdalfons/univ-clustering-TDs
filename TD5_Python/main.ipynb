# CAH

```python
# Having folowing matrix we will work in Desimulation of features
correlation
```

```python
# Desimulation of correlation matrix
distance_matrix = np.sqrt(1 - correlation**2)
distance_matrix
```

```python
# We get just the superior triangle of the matrix
from scipy.spatial.distance import squareform
squareform(distance_matrix)
```

```python
# Now create the clustering method.

from scipy.cluster.hierarchy import ward
cah = ward(squareform(distance_matrix))
```

```python
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
plt.title('Dendrogram')

dendrogram(cah, labels=X.columns, orientation='left', leaf_font_size=10, color_threshold = 0.5)
```
